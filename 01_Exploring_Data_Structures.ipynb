```python
#enable R integration
%load_ext rpy2.ipython

import scanpy as sc
import anndata2ri
from rpy2.robjects import r

#Load the .h5 count matrix file downloaded by postBuild
adata = sc.read_10x_h5("data_pbmc_3k.h5")
adata.var_names_make_unique()
```

```python
# Activate the conversion bridge
import anndata2ri
from rpy2.robjects import r, conversion, default_converter

#This assigns the Python adata object to an R variable named sce_obj
with conversion.localconverter(default_converter + anndata2ri.converter):
    r.assign("sce_obj", adata)
```
